```python
import matplotlib.pyplot as plt
categories=["electronics","clothing","home appliances","groceries"]
sales=[100000,200000,500000,56754]
plt.bar(categories,sales,color=['pink','red','black','blue'])
plt.title("monthly sales by product category")
plt.xlabel("Categories")
plt.ylabel("Sales")
plt.show()
```
